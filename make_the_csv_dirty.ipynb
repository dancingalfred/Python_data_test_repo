```python
import pandas as pd
df = pd.read_csv('8f7615b1-f6ed-41e6-871c-28129b7e9afa.csv')
df.head()
```

```python
#change all "-" in the date column to "/"
df['date'] = df['date'].str.replace('-', '/')
```

```python
#Add a new column named "submited by" to the start of the dataframe at random from list of names
list_of_names = ['Robert Johansson', 'Jenny Hassan', 'Adam Smith', 'Felicia Toll', 'Refik Bogovic']
import random
for i in range(0, len(df)):
    df.loc[i, 'submited by'] = random.choice(list_of_names)
```

```python
#Change all valuea in column mobil_2_lat to string
df['mobil_2_lat'] = df['mobil_2_lat'].astype(str)
```

```python
#change the value on row 3 in the column lejonet_temp to string and add "°C" to the end of the value
df.loc[3, 'lejonet_temp'] = str(df.loc[3, 'lejonet_temp']) + "°C"
```

```python
df.head()
```

```python
#change all the "." in the column lejonet_airpressure to ","   
df['lejonet_airpressure'] = df['lejonet_airpressure'].astype(str)
df['lejonet_airpressure'] = df['lejonet_airpressure'].str.replace('.', ',')
```

```python
#add a copy of the column femman_lat to the end of the dataframe
df['femman_lat2'] = df['femman_lat']
```

```python
#save the df as a new csv file called dirty_data.csv
df.to_csv('dirty_data.csv', index=False)
```
